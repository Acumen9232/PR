```python
import sympy as sm
x, y = sm.symbols('x,y')

f0=(x-10)*(x-1)*(x+12)*(y-10)*(y-1)*(y+12)
f1=f0.diff(x)
f2=f0.diff(y)
eq1=sm.Eq(f1,0)
eq2=sm.Eq(f2,0)

sol=sm.solve([eq1,eq2],x,y,
            dict=True,
            #simplify=True
            )
```

```python
sol
```

```python
def f(x,y): 
    zx=  (x-10)*(x-1)*(x+12) 
    zy=  (y-10)*(y-1)*(y+12)
    z= zx*zy 
    return z
z= f(x,y)
ym, xm= np.mgrid[-12:10.1:.1, -12:10.1:.1] 
zm= f(xm,ym)   
ax= pl.axes(projection='3d',  
            xlabel= 'x',  
            ylabel= 'y',  
            title= 'z= f(x,y)') 
ax.plot_surface(xm,ym,zm,  
            rstride= 5,  
            cstride= 5, 
            cmap=     'rainbow',   
            edgecolor= 'gray')
```
